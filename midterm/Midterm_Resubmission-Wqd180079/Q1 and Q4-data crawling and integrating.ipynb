## Exploratory Data analysis

```python
moving_avg =  df['Adj Close'].rolling(window=100).mean()

%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import style

# Adjusting the size of matplotlib
import matplotlib as mpl
mpl.rc('figure', figsize=(8, 7))
mpl.__version__

# Adjusting the style of matplotlib
plt.style.use( 'grayscale')

df["Adj Close"].plot(label='AAPL')
moving_avg.plot(label='moving average')
plt.legend()
```

```python
#correlation with other companies
dfcomp = web.DataReader(['AAPL', 'SNE', 'AMZN', 'IBM', 'MSFT','F','GOOG','GE'],'yahoo',start=start,end=end)['Adj Close']
```

```python
print(dfcomp.columns)
```

Percentage change finds how much the price changes compared to the previous day which defines returns and the correlation shows whether the returns are affected by other stocks’ returns,

Below is to analyse the competition by running the percentage change and correlation function,- 

```python
retscomp = dfcomp.pct_change()

corr = retscomp.corr()
```

```python
#Let’s plot Apple and GE with ScatterPlot to view their return distributions.
plt.style.use('seaborn-talk')
plt.scatter(retscomp.AAPL,retscomp.GE)
plt.xlabel("Returns AAPL")
plt.ylabel("Returns GE")
```

### From the above we can see only the relation between GE and apple stocks where it shows that except the outliers the relation is slight positive among GE returns and Apple returns.It seems the higher the Apple returns, the higher GE returns as well for most cases.
